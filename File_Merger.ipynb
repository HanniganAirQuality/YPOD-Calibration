```python
import csv
from pathlib import Path

save_dir = Path("/Users/Chiara/Documents/Documents_Chiara_MacBook_Air/Research/YPOD Calibration/Spr 26 Cal") # directory to save and get files from
pods = ["T2"] # name of the pods that need to be merged, code will not merge between pods, but all file with the same pod name specified below
base_files = ["YPOD{pod}_2025_12_18.CSV", "YPOD{pod}_2025_12_20.CSV"] # filenames of the files to merge, needs to be in order of merge

for pod in pods:
    input_files = [save_dir / name.format(pod = pod) for name in base_files] # in
    output_file = save_dir / base_files[0] # output file will have the same filename as the first file
    existing_files = [f for f in input_files if f.exists()]
    if not existing_files:
        print(f"Skipping {pod}: no files found")
        continue

    with open(output_file, "w", newline = "") as f_out:
        writer = csv.writer(f_out)
        for i, file in enumerate(existing_files):
            with open(file, "r", newline = "") as f_in:
                reader = csv.reader(f_in)  # add delimiter = if needed for tab or space delimiters
                if i > 0:
                    next(reader, None)  # skip header on all but first file
                for row in reader:
                    writer.writerow(row)
            print(f"Added {file.name}")

    print(f"Created {output_file.name}")
```
